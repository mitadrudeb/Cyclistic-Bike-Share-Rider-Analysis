# Cyclistic Bike-Share Analysis - Key Findings & Recommendations

## Executive Summary

I examined 4.86 million bike rides from 2024 to understand behavioral differences between annual members and casual riders. My goal was to identify insights that could help convert casual riders into annual members.

---

## My Key Findings

### Finding 1: Usage Volume
- **Members account for 67.4% of total rides** (3.27 million rides)
- **Casual riders account for 32.6%** (1.58 million rides)
- Despite being the minority, I found that casual riders represent a significant conversion opportunity

### Finding 2: Ride Duration
- **Casual riders take longer trips**: Average 10.61 minutes vs 9.23 minutes for members
- **Casual riders use bikes for leisure**: Longer rides suggest recreational use
- **Members use bikes for commuting**: Shorter, more efficient trips

### Finding 3: Weekly Patterns - THE BIGGEST DIFFERENCE
- **Members ride primarily on weekdays**: 76.7% of member rides occur Monday-Friday
- **Casual riders are more balanced**: 65.1% weekdays, 34.9% weekends
- **Member behavior suggests commuting**: Peak days are Wednesday and Thursday
- **Casual behavior suggests leisure**: Peak days are Saturday and Sunday

### Finding 4: Daily Timing
- **Members show clear commute patterns**: Two peaks at 7-8 AM and 4-6 PM
- **Casual riders peak in afternoons/evenings only**: 4-6 PM (no morning commute peak)
- This confirms my hypothesis that members use bikes for work commutes, while casual riders use them for leisure

### Finding 5: Seasonal Trends
- **Casual usage is highly seasonal**: 68.8% of casual rides occur May-September
- **Member usage is more consistent**: 56.6% of member rides in May-September
- **Winter drop-off is steeper for casual riders**: Only 10.7% ride in winter months
- Members maintain year-round usage, suggesting they use bikes as necessary transportation

### Finding 6: Bike Type Preferences
- **Both groups prefer electric bikes**: ~52-55% of rides
- **Casual riders use electric scooters more**: 8.49% vs -0.22% for members
- **Classic bikes used more by members**: 47.77% vs 37.86% for casual

## My Interpretation: Business Implications

### What This Means for Marketing

Based on my analysis, I've identified these distinct user profiles:

1. **Casual riders are recreational users**
   - They ride for fun, exercise, or tourism
   - They ride longer distances
   - They prefer weekends and summer months

2. **Members are commuters**
   - They use bikes as transportation to/from work
   - They ride consistently year-round
   - They follow predictable weekday patterns

3. **The Conversion Challenge**
   - Casual riders need to see value beyond weekend recreation
   - They need reasons to ride year-round
   - They need to consider bikes as a viable commute option

---



## My Top 3 Recommendations

### Recommendation 1: Weekend-to-Weekday Campaign
**Target**: Casual riders who currently only ride on weekends

**My Strategy**: 
- Launch a "Commute Challenge" campaign encouraging casual riders to try biking to work
- Offer a promotional rate: "Try membership free for first month"
- Highlight environmental and health benefits of bike commuting
- Create route planning tools showing bike paths to major business districts

**Why I believe this will work**: 34.9% of casual rides already happen on weekends, showing regular engagement. We need to extend that to weekdays.

**Expected Impact**: Convert casual riders who live near work into regular commuters


### Recommendation 2: Seasonal Membership Tier
**Target**: Casual riders who only ride in summer months

**My Strategy**:
- Create a "May-September Membership" at a reduced annual rate
- Market it as a "Summer Commuter Pass"
- Include perks: priority bike access, no unlock fees, extended ride times
- Promote heavily in April (before peak season starts)

**Why I believe this will work**: 68.8% of casual rides occur May-September. Capturing these riders for 5 months is better than losing them entirely.

**Expected Impact**: Convert seasonal recreational riders into seasonal members, with potential for year-round conversion later


### Recommendation 3: Digital Marketing - Targeted Hours
**Target**: Casual riders who already ride during commute hours

**My Strategy**:
- Run digital ads (social media, mobile app) during 4-6 PM peak hours
- Show real-time savings: "You could save $X per month with membership"
- Highlight member benefits at time of ride completion
- Send in-app notifications after casual rides: "Ride like this 3x/week? Membership saves you money"

**Why I believe this will work**: Casual riders already peak at 4-6 PM. They're already using the service during commute times - they just need to see the value of membership.

**Expected Impact**: Convert casual riders who use bikes regularly but haven't done the cost-benefit analysis
